```python
import torch
from transformers import AutoModel

# Load MODEL-2 model and print its parameters
gpt2a_model = AutoModel.from_pretrained('gpt2-large')
print("1 Model Parameters:\n")
for name, param in gpt2a_model.named_parameters():
    print(f"{name}: {param}\n")

# Load MODEL 2 model and print its parameters
gpt2b_model = AutoModel.from_pretrained('Celestinian/TopicGPT')
print("\n2 Model Parameters:\n")
for name, param in gpt2b_model.named_parameters():
    print(f"{name}: {param}\n")
```

```python
from transformers import AutoModelWithLMHead

# Define alpha, a value between 0 and 1 that controls the bias towards one of the models
alpha = 1

# Combine the parameters of gpt2a_model and gpt2b_model to create a new model3
model3 = AutoModel.from_pretrained('gpt2-large')
num_merged = 0
total_tensors = len(model3.state_dict())
for name, param in model3.named_parameters():
    if name in gpt2a_model.state_dict() and name in gpt2b_model.state_dict():
        param_a = gpt2a_model.state_dict()[name]
        param_b = gpt2b_model.state_dict()[name]
        param.data = param_a + (param_b - param_a) * alpha
        num_merged += 1
print(f"Merged {num_merged} out of {total_tensors} tensors")

# Add a language modeling head to the model
model3_lm = AutoModelWithLMHead.from_pretrained('gpt2-large')

# Copy the weights from model3 to the language modeling head
model3_lm.transformer.load_state_dict(model3.state_dict(), strict=False)
```

```python
# Instantiate the tokenizer
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2-large")

# Encode the input prompt
input_prompt = "At Sivershchyna and Slobozhanschyna directions Russian army shelled Khrinivka, Zarichchya, Yeline, Tymonovychi, Leonivka, Krasnyy Khutir of Chernihiv region; Ukrayinske, Seredyna-Buda Stukalivka, Iskryskivschyna, Shpyl, Volfyne, Yunakivka and Mohrytsya of Sumy region, also Chervona Zorya, Hoptivka, Krasne, Ternova, Vovchansk, Vovchanski Khutory and Nesterne of Kharkiv region. Russian aviation conducted airstrikes at Halahanivka and Leonivka, - General Staff of Armed Forces of Ukraine says in the morning report #CONTEXT#"
input_ids = tokenizer.encode(input_prompt, return_tensors='pt')

# Generate the output text
output_ids = model3_lm.generate(input_ids, 
                                min_length=20, 
                                max_length=200, 
                                do_sample=True, 
                                temperature=0.2, 
                                no_repeat_ngram_size=3)

# Decode the output text
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Print the output text
print(output_text)
```
